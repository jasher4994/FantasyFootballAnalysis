```python
#Calcualte ROI 
#Calculate ROM (scatter them off each other)

def graph_roi(data, position, min_minutes):
    data = data[data.minutes > min_minutes]
    data = data[data.position == position]
    positions = ["inplace", "goalkeeper", "defender", "midfielder", "foreward"]
    plt.figure(figsize = (12,8))
    plt.scatter(data["ROI"], data["now_cost"])
    plt.xlabel("ROI")
    plt.ylabel("Price")
    plt.title("ROI v Ppositions[position]")
    
    return "the graph below shows the ROI v price of {}s with a minimum minutes of {}".format(positions[position], min_minutes)

graph_roi(data, 3, 1000)
    
```
